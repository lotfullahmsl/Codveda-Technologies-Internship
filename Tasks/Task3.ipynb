# Task: Logistic Regression for Binary Classification

## i am using Churn Prediction to do this Task

```python
import pandas as pd

# load data sets
train_df = pd.read_csv('./Data/Churn_Prdiction_Data/churn-bigml-80.csv')
test_df = pd.read_csv('./Data/Churn_Prdiction_Data/churn-bigml-20.csv')

```

### Preprocessing

```python
# 1. Train Data Missing Values
train_df.isnull().sum().sort_values(ascending=False) # no missing values

# 1.1 Test Data Missing Values
test_df.isnull().sum().sort_values(ascending=False) # no missing values

# 2. drop unnecessary columns

# train_df = train_df.drop(columns=['State', 'Area code', 'Account length'])
# test_df = test_df.drop(columns=['State', 'Area code'])
# test_df = test_df.drop(columns=['Account length'])

# 3. Encoding the colums
from sklearn.preprocessing import LabelEncoder
Encoder = LabelEncoder()

train_df['International plan'] = Encoder.fit_transform(train_df['International plan'])
train_df['Voice mail plan'] = Encoder.fit_transform(train_df['Voice mail plan'])

test_df['International plan'] = Encoder.fit_transform(test_df['International plan'])
test_df['Voice mail plan'] = Encoder.fit_transform(test_df['Voice mail plan'])

# 4. Feature scaling
# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# numeric_cols = ['Number vmail messages', 'Total day minutes', 'Total day calls', 'Total day charge','Total eve charge','Total eve minutes','Total eve calls', 'Total night charge', 'Total night minutes' , 'Total night minutes', 'Total night calls', 'Total intl minutes', 'Total intl calls','Total intl charge', 'Customer service calls']
# train_df[numeric_cols] = scaler.fit_transform(train_df[numeric_cols])
# test_df[numeric_cols] = scaler.transform(test_df[numeric_cols])

test_df.head()

# 5. No need to train, test and split because already the files are separate
```

### Model Training

```python
X_train = train_df.drop(columns=['Churn'])
y_train = train_df['Churn']

X_test = test_df.drop(columns=['Churn'])
y_test = test_df['Churn']

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

### Model Evaluation

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score

y_pred = model.predict(X_test)

print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
```
