### Task: Build a Simple Linear Regression Mode

```python
import pandas as pd
import numpy as np

df = pd.read_csv('./Data/housing.csv')
```

```python
# Data Preprocessing
df.head()

# 1. drop unused colums
df = df.drop(['ocean_proximity'], axis=1)
df = df.drop(['longitude', 'latitude'], axis=1)

# 2. checking for missing value
df.isnull().sum().sort_values(ascending=False)

# 3. imputing the missing values
df['total_bedrooms'] = df['total_bedrooms'].fillna(df['total_bedrooms'].median())

# 4. feature scalling 
from sklearn.preprocessing import StandardScaler

StandardScaler = StandardScaler()

df[['total_bedrooms']] = StandardScaler.fit_transform(df[['total_bedrooms']])
df[['population']] = StandardScaler.fit_transform(df[['population']])
df[['households']] = StandardScaler.fit_transform(df[['households']])
df[['median_income']] = StandardScaler.fit_transform(df[['median_income']])
df[['median_house_value']] = StandardScaler.fit_transform(df[['median_house_value']])
df[['total_rooms']] = StandardScaler.fit_transform(df[['total_rooms']])
df[['housing_median_age']] = StandardScaler.fit_transform(df[['housing_median_age']])

```
